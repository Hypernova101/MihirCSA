# Popcorn Hack

```java
class Bro {
    String name;
    static String catchphrase = "Sup bro?";
    
    public Bro(String name) {
        this.name = name;
    }
    
    public void sayHi() {
        System.out.println("Hey, I'm " + name);
    }
    
    public static void sayCatchphrase() {
        System.out.println(catchphrase);
    }
    
    public static void main(String[] args) {
        sayCatchphrase();

        Bro alex = new Bro("Alex");
        alex.sayHi();

        alex.sayCatchphrase();
    }
}

Bro.main(null);
```

# Homework Hack

```java
class Warrior {
    String name;
    int power;
    int health;
    String weapon;
    
    static double fightDuration = 0.0;
    static int totalBattles = 0;
    
    public Warrior(String name, int power, int health, String weapon) {
        this.name = name;
        this.power = power;
        this.health = health;
        this.weapon = weapon;
    }
    
    public void attack(Warrior opponent) {
        System.out.println(name + " attacks " + opponent.name + " with " + weapon + "!");
        opponent.health -= this.power;
        if (opponent.health < 0) {
            opponent.health = 0;
        }
        System.out.println(opponent.name + " takes " + power + " damage!");
    }
    
    public void printStatus() {
        System.out.println("--- " + name + "'s Status ---");
        System.out.println("Health: " + health);
        System.out.println("Power: " + power);
        System.out.println("Weapon: " + weapon);
        System.out.println("Status: " + (health > 0 ? "Still fighting!" : "Defeated!"));
        System.out.println();
    }
    
    public void heal(int amount) {
        health += amount;
        System.out.println(name + " heals for " + amount + " HP! Current health: " + health);
    }
    
    public static int strongerFighter(Warrior w1, Warrior w2) {
        if (w1.power > w2.power) {
            System.out.println(w1.name + " is the stronger fighter with " + w1.power + " power!");
            return 1;
        } else if (w2.power > w1.power) {
            System.out.println(w2.name + " is the stronger fighter with " + w2.power + " power!");
            return 2;
        } else {
            System.out.println("Both fighters have equal power!");
            return 0;
        }
    }
    
    public static void beginBattle(Warrior w1, Warrior w2) {
        System.out.println("\n====================================");
        System.out.println("   ULTIMATE WARRIOR BATTLE");
        System.out.println("====================================");
        System.out.println(w1.name + " VS " + w2.name);
        System.out.println("====================================\n");
        
        totalBattles++;
        fightDuration = 0.0;
        
        int round = 1;
        while (w1.health > 0 && w2.health > 0) {
            System.out.println("--- ROUND " + round + " ---");
            
            w1.attack(w2);
            if (w2.health <= 0) {
                break;
            }
            
            w2.attack(w1);
            
            fightDuration += 0.5;
            round++;
            System.out.println();
        }
        
        System.out.println("\n====================================");
        if (w1.health > 0) {
            System.out.println(w1.name + " WINS THE BATTLE!");
        } else {
            System.out.println(w2.name + " WINS THE BATTLE!");
        }
        System.out.println("Battle duration: " + fightDuration + " minutes");
        System.out.println("====================================\n");
    }
    
    public static void printGameStats() {
        System.out.println("Game Statistics:");
        System.out.println("Total battles fought: " + totalBattles);
        System.out.println("Last battle duration: " + fightDuration + " minutes");
    }
    
    public static void main(String[] args) {
        Warrior thor = new Warrior("Thor", 25, 100, "Mjolnir");
        Warrior loki = new Warrior("Loki", 20, 90, "Scepter");
        
        System.out.println("WARRIORS ENTER THE ARENA\n");
        thor.printStatus();
        loki.printStatus();
        
        strongerFighter(thor, loki);
        System.out.println();
        
        loki.heal(10);
        System.out.println();
        
        beginBattle(thor, loki);
        
        System.out.println("FINAL STATUS:");
        thor.printStatus();
        loki.printStatus();
        
        printGameStats();
    }
}

Warrior.main(null);
```
