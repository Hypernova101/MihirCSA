# Popcorn Hack 1

```java
class Book {
    String title;
    int pages;

    void printInfo() {
        System.out.println("Title: " + title + ", Pages: " + pages);
    }
}

class BookDemo {
    public static void main(String[] args) {
        Book myBook = new Book();
        myBook.title = "The Great Gatsby";
        myBook.pages = 180;

        myBook.printInfo();
        
        System.out.println("\nOne more book");
        Book anotherBook = new Book();
        anotherBook.title = "1984";
        anotherBook.pages = 328;
        anotherBook.printInfo();
        
        System.out.println("\nReference vars:");
        Book book1 = new Book();
        book1.title = "Original Title";
        book1.pages = 100;
        
        Book book2 = book1;
        book2.title = "Changed Title";
        
        System.out.print("book1: ");
        book1.printInfo();
        System.out.print("book2: ");
        book2.printInfo();

        System.out.println("\nDemonstrating Object class inheritance:");
        System.out.println("Class name: " + myBook.getClass().getName());
        System.out.println("toString(): " + myBook.toString());
        System.out.println("Is Object? " + (myBook instanceof Object));
    }
}

BookDemo.main(null);
```

# Homework Hacks

```java
class Student {
    String name;
    int grade;
    int pets;
    int siblings;
    String favoriteSubject;
    
    void printInfo() {
        System.out.println("Name: " + name);
        System.out.println("Grade: " + grade);
        System.out.println("Pets: " + pets);
        System.out.println("Siblings: " + siblings);
        System.out.println("Favorite Subject: " + favoriteSubject);
        System.out.println();
    }
}

class StudentRoster {
    public static void main(String[] args) {
        Student student1 = new Student();
        student1.name = "Alex Johnson";
        student1.grade = 11;
        student1.pets = 2;
        student1.siblings = 1;
        student1.favoriteSubject = "Computer Science";
        
        Student student2 = new Student();
        student2.name = "Maria Garcia";
        student2.grade = 10;
        student2.pets = 1;
        student2.siblings = 3;
        student2.favoriteSubject = "Mathematics";
        
        Student student3 = new Student();
        student3.name = "Christopher Lee";
        student3.grade = 12;
        student3.pets = 0;
        student3.siblings = 2;
        student3.favoriteSubject = "Physics";
        
        System.out.println("Student 1:");
        student1.printInfo();
        
        System.out.println("Student 2:");
        student2.printInfo();
        
        System.out.println("Student 3:");
        student3.printInfo();
        
        System.out.println("HOMEWORK HACK 2");
        
        Student chris = student3;
        
        System.out.println("Student 3 goes by a nickname: Chris");
        System.out.println("\nAccessing through nickname reference:");
        chris.printInfo();
        
        System.out.println("--- Demonstrating Same Reference ---");
        System.out.println("Changing favorite subject through 'chris' reference...");
        chris.favoriteSubject = "Computer Science";
        
        System.out.println("\nChecking through original 'student3' reference:");
        System.out.println("student3's favorite subject: " + student3.favoriteSubject);
        System.out.println("\nChecking through 'chris' reference:");
        System.out.println("chris's favorite subject: " + chris.favoriteSubject);
        
        System.out.println("\nBoth show the same change because they reference the same object!");
        
        System.out.println("\n--- Complete Info Through Both References ---");
        System.out.println("Through student3:");
        student3.printInfo();
        
        System.out.println("Through chris (nickname):");
        chris.printInfo();
    }
}

StudentRoster.main(null);
```
