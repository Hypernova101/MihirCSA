# Popcorn Hack 1

```java
public class PowerLevelCalculator {
    
    public static void main(String[] args) {

        int basePower = 100;
        
        int level = 10;
        
        double finalPower = basePower * Math.pow(1.2, level);
        
        System.out.println("Level: " + level);
        System.out.println("Base Power: " + basePower);
        System.out.println("Final Power: " + finalPower);
        System.out.println("Final Power (rounded): " + Math.round(finalPower));
        
        System.out.println("\nPower Growth Chart:");
        for (int i = 1; i <= 15; i += 2) {
            double power = basePower * Math.pow(1.2, i);
            System.out.println("  Level " + i + ": " + Math.round(power));
        }
    }
}

PowerLevelCalculator.main(null);
```

```java
public class LootDropSimulator {
    
    public static void main(String[] args) {        
        int rarityRoll = (int)(Math.random() * 100) + 1;
        
        String itemRarity;
        int goldValue;
        
        if (rarityRoll >= 1 && rarityRoll <= 60) {
            itemRarity = "COMMON";
            goldValue = (int)(Math.random() * (30 - 10 + 1)) + 10;
            
        } else if (rarityRoll >= 61 && rarityRoll <= 85) {
            itemRarity = "RARE";
            goldValue = (int)(Math.random() * (70 - 31 + 1)) + 31;
            
        } else {
            itemRarity = "LEGENDARY";
            goldValue = (int)(Math.random() * (100 - 71 + 1)) + 71;
        }
        
        System.out.println("Loot Drop");
        System.out.println("Rarity: " + rarityRoll);
        System.out.println("You got: " + itemRarity + " item");
        System.out.println("Gold Value: " + goldValue);
        
        if (itemRarity.equals("LEGENDARY")) {
            System.out.println("Wow");
        } else if (itemRarity.equals("RARE")) {
            System.out.println("Nice job");
        }
        
        for (int i = 1; i <= 3; i++) {
            System.out.println("\nDrop #" + i + ":");
            simulateDrop();
        }
    }
    
    public static void simulateDrop() {
        int rarityRoll = (int)(Math.random() * 100) + 1;
        String itemRarity;
        int goldValue;
        
        if (rarityRoll >= 1 && rarityRoll <= 60) {
            itemRarity = "COMMON";
            goldValue = (int)(Math.random() * (30 - 10 + 1)) + 10;
        } else if (rarityRoll >= 61 && rarityRoll <= 85) {
            itemRarity = "RARE";
            goldValue = (int)(Math.random() * (70 - 31 + 1)) + 31;
        } else {
            itemRarity = "LEGENDARY";
            goldValue = (int)(Math.random() * (100 - 71 + 1)) + 71;
        }

        System.out.println("Rarity: " + rarityRoll);
        System.out.println("You got: " + itemRarity + " item");
        System.out.println("Gold Value: " + goldValue);
    }
}

LootDropSimulator.main(null);
```

# Homework Hack

```java
public class GameStatsCalculator {
    
    public static int healthDifference(int player1Health, int player2Health) {
        int difference = player1Health - player2Health;
        return Math.abs(difference);
    }
    
    public static double calculateDamage(double baseDamage, double powerLevel) {
        double damage = baseDamage * Math.pow(1.5, powerLevel);
        return damage;
    }

    public static double findDistance(int playerX, int playerY, int enemyX, int enemyY) {
        int xDiff = enemyX - playerX;
        int yDiff = enemyY - playerY;

        double xSquared = Math.pow(xDiff, 2);
        double ySquared = Math.pow(yDiff, 2);

        double distance = Math.sqrt(xSquared + ySquared);
        
        return distance;
    }

    public static int generateLoot(int minValue, int maxValue) {
        int randomLoot = (int)(Math.random() * (maxValue - minValue + 1)) + minValue;
        return randomLoot;
    }

    public static void main(String[] args) {
        System.out.println("Game Stats");

        System.out.println("Part A");
        System.out.println("healthDifference(75, 120) = " + healthDifference(75, 120));
        System.out.println("healthDifference(100, 80) = " + healthDifference(100, 80));
        System.out.println("healthDifference(50, 50) = " + healthDifference(50, 50));

        System.out.println("Part B");
        System.out.println("calculateDamage(10.0, 2) = " + calculateDamage(10.0, 2));
        System.out.println("calculateDamage(15.0, 3) = " + calculateDamage(15.0, 3));
        System.out.println("calculateDamage(20.0, 1) = " + calculateDamage(20.0, 1));

        System.out.println("Part C");
        System.out.println("findDistance(0, 0, 3, 4) = " + findDistance(0, 0, 3, 4));
        System.out.println("findDistance(1, 1, 4, 5) = " + findDistance(1, 1, 4, 5));
        System.out.println("findDistance(0, 0, 6, 8) = " + findDistance(0, 0, 6, 8));

        System.out.println("Part D");
        System.out.println("Generating random loot (10-50):");
        for (int i = 1; i <= 5; i++) {
            System.out.println("  Drop " + i + ": " + generateLoot(10, 50) + " gold");
        }
        
        System.out.println("\nGenerating dice rolls (1-6):");
        for (int i = 1; i <= 5; i++) {
            System.out.println("  Roll " + i + ": " + generateLoot(1, 6));
        }
        
        System.out.println("\nGenerating loot with same min/max (100):");
        System.out.println("  generateLoot(100, 100) = " + generateLoot(100, 100));
        
        System.out.println("Complete game scenario:");
        int player1Health = 85;
        int player2Health = 120;
        System.out.println("Player 1 Health: " + player1Health);
        System.out.println("Player 2 Health: " + player2Health);
        System.out.println("Health Gap: " + healthDifference(player1Health, player2Health) + " HP");
        
        double baseDmg = 12.0;
        double powLvl = 2;
        System.out.println("\nPlayer attacks with base damage " + baseDmg + " at power level " + powLvl);
        System.out.println("Total Damage: " + calculateDamage(baseDmg, powLvl));
        
        int px = 5, py = 5, ex = 8, ey = 9;
        System.out.println("\nPlayer at (" + px + ", " + py + ")");
        System.out.println("Enemy at (" + ex + ", " + ey + ")");
        System.out.println("Distance: " + findDistance(px, py, ex, ey) + " units");
        
        System.out.println("\nEnemy defeated! Loot drop:");
        System.out.println("Gold earned: " + generateLoot(25, 75) + " gold");
    }
}

GameStatsCalculator.main(null);
```
