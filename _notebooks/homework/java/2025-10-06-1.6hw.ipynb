# Popcorn Hack 1

```java
// ============================================
// POPCORN HACK #1: Transform the code
// ============================================

public class PopcornHack1 {
    public static void main(String[] args) {
        System.out.println("=== POPCORN HACK #1 ===\n");
        
        int playerScore = 1000;
        int playerHealth = 100;
        int enemiesDefeated = 0;
        
        System.out.println("Starting stats:");
        System.out.println("Score: " + playerScore + ", Health: " + playerHealth + ", Enemies: " + enemiesDefeated);
        
        // Player defeats an enemy worth 250 points
        playerScore += 250;
        System.out.println("\nDefeated enemy! Score: " + playerScore);
        
        // Player takes 15 damage
        playerHealth -= 15;
        System.out.println("Took damage! Health: " + playerHealth);
        
        // Enemy count goes up
        enemiesDefeated++;
        System.out.println("Enemy defeated count: " + enemiesDefeated);
        
        // Boss battle: double the current score!
        playerScore *= 2;
        System.out.println("Boss defeated - Score doubled! Score: " + playerScore);
        
        // Healing potion restores health to 80% of current
        playerHealth *= 4;
        playerHealth /= 5;  // 4/5 = 0.8, using integer math
        System.out.println("Used healing potion (80% restore)! Health: " + playerHealth);
        
        System.out.println("\nFinal stats:");
        System.out.println("Score: " + playerScore + ", Health: " + playerHealth + ", Enemies: " + enemiesDefeated);
    }
}

PopcornHack1.main(null);
```

# Popcorn Hack 2

```java
// ============================================
// POPCORN HACK #2: Three compound operators
// ============================================

public class PopcornHack2 {
    public static void main(String[] args) {
        System.out.println("=== POPCORN HACK #2 ===\n");
        
        int score = 100;
        System.out.println("Starting score: " + score);
        
        // Deduct points for a wrong answer
        score -= 15;
        System.out.println("Wrong answer penalty (-15): " + score);
        
        // Double the score with a power-up
        score *= 2;
        System.out.println("Power-up activated (x2): " + score);
        
        // Find remainder after dividing by 7
        score %= 7;
        System.out.println("Remainder after dividing by 7: " + score);
        
        System.out.println("\nFinal score: " + score);
    }
}

PopcornHack2.main(null);
```

# Homework Hack Option C

```java
// ============================================
// HOMEWORK: Option C - Fitness Progress Tracker
// ============================================

public class FitnessTracker {
    public static void main(String[] args) {
        System.out.println("=== FITNESS PROGRESS TRACKER ===\n");
        
        // Initialize variables
        int totalCalories = 2000;  // Starting daily calorie balance
        int workoutDays = 0;       // Days exercised this week
        int stepCount = 5000;      // Steps taken today
        int weeklyGoal = 7;        // Goal: workout 7 days
        int progress = 0;          // Progress tracker
        
        System.out.println("Starting metrics:");
        System.out.println("Total Calories: " + totalCalories);
        System.out.println("Workout Days: " + workoutDays);
        System.out.println("Step Count: " + stepCount + "\n");
        
        // Morning: Breakfast calories
        totalCalories += 450;
        System.out.println("Ate breakfast!");
        System.out.println("Total Calories: " + totalCalories + " (+450 from breakfast)\n");
        
        // Lunch time
        totalCalories += 600;
        System.out.println("Ate lunch!");
        System.out.println("Total Calories: " + totalCalories + " (+600 from lunch)\n");
        
        // Afternoon workout session
        totalCalories -= 350;
        workoutDays++;
        System.out.println("Completed afternoon workout!");
        System.out.println("Total Calories: " + totalCalories + " (-350 burned from workout)");
        System.out.println("Workout Days: " + workoutDays + " (Great job!)\n");
        
        // Evening walk - very active!
        stepCount *= 2;
        System.out.println("Took a long evening walk!");
        System.out.println("Step Count: " + stepCount + " (doubled from active walking!)\n");
        
        // Dinner
        totalCalories += 700;
        System.out.println("Ate dinner!");
        System.out.println("Total Calories: " + totalCalories + " (+700 from dinner)\n");
        
        // Evening workout
        totalCalories -= 300;
        workoutDays++;
        System.out.println("Evening cardio session!");
        System.out.println("Total Calories: " + totalCalories + " (-300 burned)");
        System.out.println("Workout Days: " + workoutDays + "\n");
        
        // Calculate weekly average steps (assuming 7 days)
        int avgSteps = stepCount / 7;
        System.out.println("Calculating weekly average steps...");
        System.out.println("Average daily steps (7-day): " + avgSteps + "\n");
        
        // Track progress toward weekly goal
        progress = workoutDays % weeklyGoal;
        System.out.println("Checking weekly workout goal progress...");
        System.out.println("Progress: " + progress + " (days toward next weekly cycle)\n");
        
        // Final summary
        System.out.println("=== DAILY SUMMARY ===");
        System.out.println("Net Calories: " + totalCalories);
        System.out.println("Workout Days This Week: " + workoutDays);
        System.out.println("Total Steps Today: " + stepCount);
        System.out.println("Weekly Goal Progress: " + progress + " days");
        
        // Motivational message
        if (totalCalories < 2500) {
            System.out.println("\nGreat calorie management today!");
        } else {
            System.out.println("\nSlightly over calorie target, but you worked out hard!");
        }
        
        if (workoutDays >= 2) {
            System.out.println("Awesome workout consistency!");
        }
        
        System.out.println("\n=== TRACKER COMPLETE ===");
    }
}

FitnessTracker.main(null);
```
